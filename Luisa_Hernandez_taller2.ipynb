## Visualización para ciencia de datos con python

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g = nx.read_edgelist('bn-mouse_visual-cortex_1.edges')

print("Número de nodos:", g.number_of_nodes())
print("Número de nodos:", g.number_of_edges())
```

```python
sorted(list(g.nodes))
```

```python
sorted(list(g.edges))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(50,20))
plt.subplot(121)
nx.draw(g, with_labels=True,
        node_color='blue',
        node_size=1500, 
        width=3,
        edge_cmap=plt.cm.Wistia)
plt.title("Mouse cortex")
```

## ¿Qué?
###  En el grafico observamos nodos los cuales son atributos cuantitativos  que van del 0 al 28 ,con un total de 29 nudos ,tambien podemos observar enlaces que son las relaciones que existen entre los nodos con un total de 44 enlaces,el tipo de dataset es un networks.
